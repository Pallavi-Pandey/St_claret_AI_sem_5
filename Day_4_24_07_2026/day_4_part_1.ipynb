# Linear Regression Models - Machine Learning

This notebook demonstrates **Simple Linear Regression** (1 input feature) and **Multiple Linear Regression** (2 and 4 input features) using `scikit-learn` to solve sequence prediction problems.

## Problem 1: Simple Linear Regression

### Problem Statement
Predict the value of $y$ when $x = 10$, given the data points:
- $x = 1 \implies y = 3$
- $x = 2 \implies y = 5$
- $x = 3 \implies y = 7$

### Analytical Solution
Observing the pattern:
- $y = 2(1) + 1 = 3$
- $y = 2(2) + 1 = 5$
- $y = 2(3) + 1 = 7$

The linear relationship is: **$y = 2x + 1$**

When $x = 10 \implies y = 2(10) + 1 = \mathbf{21}$.

```python
# Problem 1: Simple Linear Regression Implementation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Define dataset
X_train_1 = np.array([[1], [2], [3]])
y_train_1 = np.array([3, 5, 7])

# 2. Initialize and fit the Linear Regression model
model_1 = LinearRegression()
model_1.fit(X_train_1, y_train_1)

# 3. Display learned slope (m) and intercept (c)
m = model_1.coef_[0]
c = model_1.intercept_
print(f"Learned Linear Equation: y = {m:.2f} * x + {c:.2f}")

# 4. Predict y when x = 10
X_test_1 = np.array([[10]])
y_pred_1 = model_1.predict(X_test_1)[0]
print(f"Predicted value of y when x = 10: {y_pred_1:.2f}")
```

```python
# Visualization for Problem 1
plt.figure(figsize=(8, 5))

# Plot regression line
x_line = np.linspace(0, 11, 100).reshape(-1, 1)
y_line = model_1.predict(x_line)
plt.plot(x_line, y_line, color='blue', label=f'Regression Line (y = {m:.1f}x + {c:.1f})')

# Plot training data points
plt.scatter(X_train_1, y_train_1, color='green', s=100, zorder=5, label='Training Points')

# Plot prediction target point
plt.scatter(10, y_pred_1, color='red', marker='*', s=250, zorder=6, label=f'Prediction at x=10 (y={y_pred_1:.0f})')

plt.title('Problem 1: Simple Linear Regression Model', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

## Problem 2: Multiple Linear Regression (2 Features)

### Problem Statement
Predict the value of $z$ when $x = 10$ and $y = 20$, given the data points:
- Sample 1: $x = 2, y = 2 \implies z = 10$
- Sample 2: $x = 3, y = 4 \implies z = 17$

### Analytical Solution
Using a multiple linear model: **$z = w_1 x + w_2 y + b$**

Substituting samples:
1) $2 w_1 + 2 w_2 + b = 10$
2) $3 w_1 + 4 w_2 + b = 17$

Setting intercept $b = 0$:
- $w_1 + w_2 = 5$
- $3 w_1 + 4 w_2 = 17 \implies w_1 = 3, w_2 = 2$

The linear relationship is: **$z = 3x + 2y$**

When $x = 10$ and $y = 20 \implies z = 3(10) + 2(20) = 30 + 40 = \mathbf{70}$.

```python
# Problem 2: Multiple Linear Regression Implementation

# 1. Define dataset with 2 features [x, y] and target z
X_train_2 = np.array([[2, 2], [3, 4]])
z_train_2 = np.array([10, 17])

# 2. Initialize and fit Multiple Linear Regression model
model_2 = LinearRegression()
model_2.fit(X_train_2, z_train_2)

# 3. Extract coefficients and intercept
w1, w2 = model_2.coef_
b = model_2.intercept_
print(f"Learned Equation: z = {w1:.4f} * x + {w2:.4f} * y + {b:.4f}")

# 4. Predict z when x = 10 and y = 20
X_test_2 = np.array([[10, 20]])
z_pred_2 = model_2.predict(X_test_2)[0]
print(f"Predicted value of z when x = 10 and y = 20: {z_pred_2:.4f}")

# 5. Verification on training set
print("\nVerification on Training Data:")
for (x_val, y_val), z_val in zip(X_train_2, z_train_2):
    pred = model_2.predict([[x_val, y_val]])[0]
    print(f"  Input (x={x_val}, y={y_val}) -> Actual z={z_val}, Model Output z={pred:.4f}")
```

```python
# 3D Surface Visualization for Problem 2
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Training points
ax.scatter(X_train_2[:, 0], X_train_2[:, 1], z_train_2, color='green', s=100, label='Training Points')

# Target prediction point
ax.scatter(10, 20, z_pred_2, color='red', marker='*', s=250, label=f'Prediction (10, 20, z={z_pred_2:.1f})')

# Create mesh grid surface for prediction surface
x_surf = np.linspace(0, 12, 20)
y_surf = np.linspace(0, 22, 20)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)
z_surf = model_2.predict(np.c_[x_surf.ravel(), y_surf.ravel()]).reshape(x_surf.shape)

ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.4, cmap='coolwarm')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.set_title('Problem 2: Multiple Linear Regression Plane', fontsize=14)
ax.legend()
plt.show()
```
